```python
# curl -X 'POST' http://127.0.0.1:1337/fileCloud/login -H "Content-Type: application/json" -H "Authorization: Bearer ${}"" -d '{"username":"ESB", "password":"test"}'
def authorize():
    url = "http://127.0.0.1:1337/fileCloud/login"
    headers = {'Content-type': 'application/json', 'Accept': 'application/json'}
    data = {"username":"ESB", "password":"test"}
    r = requests.post(url, data=json.dumps(data), headers=headers)
    token = json.loads((r.content).decode('utf-8'))['access_token']
    print (r.request.body) 
    return token
```

```python
def download(token, data):
    url = "http://127.0.0.1:1337/fileCloud/download"
    headers = {'Accept': 'application/octet-stream', 'Authorization': 'Bearer {}'.format(token) }   
    r = requests.post(url, headers=headers, data=data)
    print(r.request.body)
    return r
```

```python
#'Content-type': 'multipart/form-data'
def upload(token, file, data):
    url = "http://127.0.0.1:1337/fileCloud/upload"
    headers = {'Accept': 'application/json', 'Authorization': 'Bearer {}'.format(token) }     
    r = requests.post(url, headers=headers, data=data, files={"File":('test.jpg',file,'application-type')})
    print (r.request.body)    
    return r
```
